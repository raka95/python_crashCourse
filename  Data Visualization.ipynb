alpha : for transparency
normed : if we want different datasets to be ploted with different data ponits
density : boolean, optional
If True, the first element of the return tuple will be the counts normalized to form a probability density, i.e., the area (or integral) under the histogram will sum to 1. This is achieved by dividing the count by the number of observations times the bin width and not dividing by the total number of observations. If stacked is also True, the sum of the histograms is normalized to 1.

Default is None for both normed and density. If either is set, then that value will be used. If neither are set, then the args will be treated as False.

If both density and normed are set an error is raised.

## V50 Box and Violin plots

```python
# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

# Import the stats library from numpy
from scipy import stats

# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that plots appear in the iPython Notebook
%matplotlib inline
```

```python
# Now we'll learn about box and violin plots
url = 'http://en.wikipedia.org/wiki/Box_plot#mediaviewer/File:Boxplot_vs_PDF.svg'

# Let's create two distributions
data1 = randn(100)
data2 = randn(100) + 2 # Off set the mean
```

```python
#sns.boxplot([data1,data2])
sns.boxplot(data=[data1,data2])
```

```python
sns.boxplot(data=[data1,data2],whis=np.inf)
#outliers included
```

```python
#sns.boxplot(data=data1,vert=False)
sns.boxplot(data=data1,orient='h')

```

```python
#normal dist.
data1=stats.norm(0,5).rvs(100)
# 2 gamma dist. concatenated together
data2=np.concatenate([stats.gamma(5).rvs(50)-1,
                    -1*stats.gamma(5).rvs(50)])
#boxplot data1 and data2
sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
#combine KDE plot with boxplot
sns.violinplot(data=[data1,data2])
```

```python
 sns.violinplot(data=data2,bw=0.01)
```

```python
#rugplot included
sns.violinplot(data1,inner='stick')
```

## V51 Reggression plots

```python
#seaborn datasets 
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#linear regression plot
sns.lmplot('total_bill','tip',tips)
```

```python
# using dictionaries
sns.lmplot('total_bill','tip',tips,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
#specifing what order fit we want if we want the linear
#linear ==> order=1 (default)
sns.lmplot('total_bill','tip',tips,order=2,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
#ignoring reggression plot and only scatter plot will be shown 
sns.lmplot('total_bill','tip',tips,fit_reg=False)
```

```python
tips.head()
```

```python
tips['tip_pect']=100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

```python
sns.lmplot('size','tip_pect',tips)
```

```python
# We can also add jitter to this

#Info link
url = "http://en.wikipedia.org/wiki/Jitter"

#plot
sns.lmplot("size", "tip_pect", tips,x_jitter=.1);
```

```python
import webbrowser
webbrowser.open(url)
```

```python
sns.lmplot('size','tip_pect',tips,x_estimator=np.mean)
```

```python
sns.lmplot('total_bill','tip_pect',tips,hue='sex',markers=['x','o'])
```

```python
sns.lmplot(x='total_bill',y='tip_pect',data=tips,hue='day')
```

```python
# Finally it should be noted that Seabron supports LOESS model fitting
url = 'http://en.wikipedia.org/wiki/Local_regression'
webbrowser.open(url)
```

```python
# lowess ==> local regression
sns.lmplot('total_bill','tip_pect',tips,
           lowess=True,line_kws={'color':'black'})
```

```python
#lower level function that lmplot is built on top 
sns.regplot('total_bill','tip_pect',tips)
#more robust
```

```python
#figure of 2 subplots
#first line allows this for us 
fig, (axis1,axis2)=plt.subplots(1,2,sharey=True)
sns.regplot('total_bill','tip_pect',tips,ax=axis1)
sns.violinplot(y=tips['tip_pect'],x=tips['size'],color='indianred',ax=axis2)
```
